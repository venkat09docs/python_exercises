##  Modifying Data: Adding and Dropping Columns and Rows

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['City'] = ['New York', 'Los Angeles', 'Chicago']
```

```python
df
```

```python
df['Age Group'] = ['Young' if age < 30 else 'Mature' for age in df['Age']]
```

```python
df
```

#### axis 0 is row and 1 is column

```python
df = df.drop('Age Group', axis=1)
```

```python
df
```

```python
df = df.drop(1)
```

```python
df
```

```python
df_copy = df.copy()
```

```python
df_copy
```

```python
df.assign(Discount=[5, 10], Total_Spend=[100, 200])
```

## Filtering Data: Masks and pandas.Series.between()

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
mask = df['Age'] > 30
```

```python
mask
```

```python
filtered_df = df[mask]
filtered_df
```

```python
df
```

```python
mask = (df['Age'] > 30) & (df['City'] == 'Houston')
filtered_df = df[mask]
filtered_df
```

```python
df
```

```python
mask = df['Age'].between(30, 40)
filtered_df = df[mask]
filtered_df
```

```python
df
```

```python
mask = df['City'].isin(['Los Angeles', 'Houston'])
filtered_df = df[mask]
filtered_df
```
