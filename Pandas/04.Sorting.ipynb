```python
sorted_df = df.sort_values(by='Age')
sorted_df
```

```python
sorted_df = df.sort_values(by='Age', ascending=False)
sorted_df
```

```python
sorted_df = df.sort_values(by='City')
sorted_df
```

```python
sorted_df = df.sort_values(by=['City', 'Age'])
sorted_df
```

```python
sorted_df = df.sort_index()
sorted_df
```

```python
# Example DataFrame: Player leaderboard with names and scores
data = {
    'Player': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Score': [150, 200, 125, 300, 175]
}
df = pd.DataFrame(data)
df
```

```python
sorted_df = df.sort_values(by='Score', ascending=False).reset_index(drop=True)
sorted_df['Rank'] = sorted_df.index + 1
sorted_df
```

## Handling Missing Data

```python
import pandas as pd
# Sample DataFrame with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, None, 35, 40],
    'City': ['New York', 'Los Angeles', None, 'Houston']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.isna()
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_cleaned = df.dropna(axis=1)
df_cleaned
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df
```

```python
df['City'] = df['City'].fillna('Unknown')
df
```

```python
df.isnull()
```

## Aggregations and Grouping Data

```python
import pandas as pd
# Sample DataFrame
data = {
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Apparel'],
    'Product': ['Laptop', 'Mouse', 'Chair', 'Table', 'Shoes'],
    'Sales': [1200, 100, 400, 300, 50]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
grouped_df = df.groupby('Category')['Sales'].sum().reset_index()
grouped_df
```

```python
grouped_df = df.groupby(['Category','Product'])['Sales'].sum()
grouped_df
```

```python
summary = df.groupby('Category')['Sales'].agg(['sum', 'count', 'mean'])
summary
```
