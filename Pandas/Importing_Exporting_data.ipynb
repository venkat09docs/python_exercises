## Importing Data: Working with CSV Files

```python
import pandas as pd
```

```python
df = pd.read_csv('model_logs.csv', encoding="utf-8")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.info()
```

```python
df['Response Time (ms)'] = df['Response Time (ms)'].astype(float)
df.info()
```

```python
df
```

```python
mask = df['Response Time (ms)'] > 50
slow_responses = df[mask]
slow_responses
```

### Export the Data to CSV File

```python


slow_responses.to_csv('slow_responses.csv', index=False)
```
